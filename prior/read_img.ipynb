#### 1.opencv

```python
import  cv2
```

```python
# 读取图片,图片的numpy是以BGR排列，不是以RGB排列的方式
#用opencv读取图片的时候，如果图片不存在，并不会报错
x1=cv2.imread('./test_img.jpg')
```

返回的数组为(高，宽，channels)

```python
x1.shape
```

```python
#将数组保存为图像
cv2.imwrite('x1.jpg',x1)
```

#### 2. keras.preprocessing.image

```python
from keras.preprocessing import image
#读取图片，图片的numpy是以RGB的形式排列
x2=image.load_img('test_img.jpg')
x2=image.img_to_array(x2)
```

```python
#(高，宽，channels)
x2.shape
```

#### 3.PIL.Image

```python
from PIL import Image
import numpy as np
x3=Image.open('test_img.jpg')
```

```python
x3
```

```python
#存储图片
# x3=np.array(x3)
x3=Image.fromarray(x3)
x3.save('x3.jpg')
```

#### 4.scipy.misc

```python
from scipy import misc
x4=misc.imread('test_img.jpg')
```

```python
#(高，宽，channels)
x4.shape
```

```python
#保存图像
misc.imsave('x4.jpg',x4)
```
