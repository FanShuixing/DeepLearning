### model: deeplab v3+ data: 遥感二分类


```python
from model import Deeplabv3
model=Deeplabv3(weights=None,input_shape=(224,224,4),classes=2,backbone='xception')
model.load_weights('./models/deeplab_v3+.hdf5')
```

```python
from keras import optimizers
# model=FCN_Vgg16_32s(input_shape=(320,320,3),classes=21)
optimizer=optimizers.RMSprop(lr=1e-5)
#可以自定义损失函数和metrics
# loss_fn=softmax_sparse_crossentropy_ignoring_last_label
# metrics=[sparse_accuracy_ignoring_last_label]
model.compile(loss='binary_crossentropy',optimizer=optimizer,metrics=['acc'])
model.summary()

```

```python
import numpy as np
import os
from keras.callbacks import ModelCheckpoint
ori_train=np.load('./data/train_image.npy')
# print(ori_train.shape)
xs_train=np.load('./data/train_image.npy').swapaxes(1,2).swapaxes(2,3)
ys_train=np.load('./data/train_annotation.npy')

xs_val=np.load('./data/validation_image.npy').swapaxes(1,2).swapaxes(2,3)
ys_val=np.load('./data/validation_annotation.npy')

# print(xs_train.shape,ys_train.shape)

if  not os.path.exists('./models'):
    os.mkdir('./models')
    
checkpoint=ModelCheckpoint('./models/deeplab_v3+.hdf5',monitor='val_acc',save_best_only=True,verbose=1)

model.fit(xs_train,ys_train,batch_size=16,epochs=20,callbacks=[checkpoint],validation_data=(xs_val,ys_val))
```
